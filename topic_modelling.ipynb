```python
import pandas as pd
import gensim
import re
from gensim.utils import simple_preprocess
from gensim.models import CoherenceModel
from nltk.corpus import stopwords
import gensim.parsing.preprocessing as gsp

# Load the CSV file into a pandas dataframe
df = pd.read_csv('filtered_6months_Whi.csv')


stop_words = gsp.STOPWORDS.union(set(['speaker', 'okay', 'like', 'thank', 'call', 'right', 'hi', 'whats', 'yes', 'call', 'yeah']))

def preprocess(text):
    result = []
    for token in simple_preprocess(text):
        if token not in stop_words:
            result.append(token)
    return result

def unicodes_remove(text):
    text = re.sub(r"(@\[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?", "", text)
    return text
```

```python
# Loop through each row in the dataframe
for index, row in df.iterrows():

    # Preprocess the text for the current row
    text = row['TEXT']
    tokens = preprocess(text)

    # Create a dictionary and corpus for the LDA model
    dictionary = gensim.corpora.Dictionary([tokens])
    corpus = [dictionary.doc2bow(tokens)]

    # Train the LDA model with 10 topics
    lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                                id2word=dictionary,
                                                num_topics=10,
                                                random_state=100,
                                                update_every=1,
                                                chunksize=100,
                                                passes=10,
                                                alpha='auto',
                                                per_word_topics=True)

    # Get the top 3 topics and their top 10 most representative words
    top_topics = lda_model.top_topics(corpus, topn=5)
    top_3_topics = top_topics[:5]

    # Print the top 3 topics and their top 10 most representative words
    for idx, topic in enumerate(top_3_topics):
        print('Row: {} \nTopic {}: \nWords: {}'.format(index, idx+1, topic[0]))
```

```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from gensim.utils import simple_preprocess
from bertopic import BERTopic
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')

# Load the CSV file into a pandas dataframe
df = pd.read_csv('filtered_6months_Whi.csv')

stop_words = stopwords.words('english')
stop_words.extend(['speaker', 'okay', 'like', 'thank', 'right', 'hi', 'forget', 'version', 'part', 'case', 'squad', 'seen','go','whats', 'yes', 'call', 'number', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine', 'address', 'said','cant', 'thing', 'emergency', 'late', 'get', 'department', 'happen', 'gonna', 'if', 'else', 'ohio', 'ahead', 'he', 'you', 'she', 'welcome', 'ahead', 'month', 'year', 'ago', 'telephone', 'car', 'paint', 'mom', 'everybody', 'yelling', 'everyone', 'great', 'kind', 'game', 'talking', 'side', 'wall', 'residence', 'outside', 'byebye', 'couldnt', 'little', 'side', 'help', 'start', 'ask', 'yeah', 'road', 'im', 'outside', 'got', 'going', 'located', 'building', 'dining', 'floor', 'room', 'hallway', 'exactly', 'called', 'day', 'might', 'guy', 'sorry'])

# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

def preprocess(text):
    result = []
    for token in simple_preprocess(text):
        if token not in stop_words:
            # Lemmatize each token
            lemma = lemmatizer.lemmatize(token)
            result.append(lemma)
    return result

def unicodes_remove(text):
    text = re.sub(r"(@\[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?", "", text)
    return text

# Create a BERTopic model
model = BERTopic(min_topic_size=5)

# Loop through each row in the dataframe
for index, row in df.iterrows():

    # Preprocess the text for the current row
    text = row['TEXT']
    text = unicodes_remove(text)
    tokens = preprocess(text)

    # Fit the BERTopic model to the tokens
    topics, probs = model.fit_transform(tokens)

    # Get the top topic ID
    top_topic = topics[0]

    # Get the top 8 most representative words for the top topic
    top_words = model.get_topic(top_topic)[:8]

    # Print the top topic and its top 8 most representative words
    print('Row: {}\n text: {} \nTop Topic: {}\nTop Words: {}\n'.format(index, tokens, top_topic, top_words))
```
